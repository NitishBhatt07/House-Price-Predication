```python
train = train.drop(['Id','Alley','FireplaceQu','PoolQC','Fence','MiscFeature'],axis=1)
```

```python
train.GarageQual = train.GarageQual.fillna(train.GarageQual.mode()[0])
```

```python
train.GarageFinish = train.GarageFinish.fillna(train.GarageFinish.mode()[0])
```

```python
train.GarageType = train.GarageType.fillna(train.GarageType.mode()[0])
```

```python
train.Electrical = train.Electrical.fillna(train.Electrical.mode()[0])
```

```python
train.BsmtFinType2 = train.BsmtFinType2.fillna(train.BsmtFinType2.mode()[0])
```

```python
train.BsmtFinType1 = train.BsmtFinType1.fillna(train.BsmtFinType1.mode()[0])
```

```python
train.BsmtExposure = train.BsmtExposure.fillna(train.BsmtExposure.mode()[0])
```

```python
train.BsmtCond = train.BsmtCond.fillna(train.BsmtCond.mode()[0])
```

```python
train.BsmtQual = train.BsmtQual.fillna(train.BsmtQual.mode()[0])
```

```python
train.MasVnrType = train.MasVnrType.fillna(train.MasVnrType.mode()[0])
```

```python
train.GarageYrBlt = train.GarageYrBlt.fillna(train.GarageYrBlt.mode()[0])
```

```python
train.MasVnrArea = train.MasVnrArea.fillna(train.MasVnrArea.mean())
```

```python
## Filling outliers LotFrontage
LotFrontage_mean = np.mean(train.LotFrontage)
LotFrontage_std = np.std(train.LotFrontage)

lower = LotFrontage_mean - 3*LotFrontage_std
upper = LotFrontage_mean + 3*LotFrontage_std

LotFrontage_out_rm = [x for x in train.LotFrontage if x >= lower and x <= upper]

rm_values = len(train.LotFrontage) - len(LotFrontage_out_rm)
for i in range(rm_values):
    LotFrontage_out_rm.append(np.median(LotFrontage_out_rm))

train['LotFrontage'] = LotFrontage_out_rm
```

```python
catagorical_features = [column for column in train.columns if train.dtypes[column] == 'object']
```

```python
data = [i for i in range(0,1460)]
catagorical_dummies = pd.DataFrame(data)
for column in catagorical_features:
    catagorical_dummies = pd.concat([catagorical_dummies,pd.get_dummies(train[column] , drop_first=True)],axis=1)
```

```python
from sklearn.model_selection import KFold,cross_val_score

def Cross_Validation(model ,X,y):
    cv = KFold(n_splits=10, random_state=1, shuffle=True)
    # evaluate model
    scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    # report performance
    print('Accuracy on Train: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
from sklearn.linear_model import LogisticRegression
Cross_Validation(LogisticRegression(),X,y)
```
